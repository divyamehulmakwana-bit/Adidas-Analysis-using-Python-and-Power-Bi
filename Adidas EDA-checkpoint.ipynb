# Exploratory Data Analysis on Adidas Sales Dataset

**Adidas Sales Dataset â€“ Summary**

This dataset contains transactional-level sales data for Adidas products across different regions and retailers. It captures both revenue and profitability details, enabling analysis from multiple perspectives (time, geography, product, retailer, and sales channel).

**Problem Statement**

Adidas, as a global sportswear brand, generates sales across multiple regions, retailers, and sales channels. However, without structured analysis, it is difficult to identify which regions, products, and retailers drive the most revenue and profit, and how different sales methods (online, offline, distributor) impact performance.

This project aims to perform Exploratory Data Analysis (EDA) on Adidas sales data to uncover trends, measure profitability, and highlight actionable insights that can guide business decisions on product focus, regional strategies, and channel optimization.

Adidas Sales Data Analysis â€“ Key Insights

The analysis of the Adidas sales dataset provides a detailed understanding of the companyâ€™s performance across products, regions, retailers, and customer segments.

Revenue & Profit Insights:

Revenue is heavily concentrated in certain products and categories, with Menâ€™s Street Footwear and Street Wear dominating sales and profitability.

Operating profit trends closely follow revenue, indicating strong margin products are also top sellers.

Significant growth is observed in certain periods, suggesting seasonal or promotional spikes, particularly in 2021.

Product & Customer Segmentation:

Sales are fairly balanced between male and female products, but Menâ€™s products contribute slightly higher revenue.

High-margin products tend to be concentrated in specific categories, offering opportunities for focused marketing and inventory strategies.

Product performance varies across regions, suggesting regional preferences or retailer specialization.

Regional & Retailer Insights:

Certain states and cities contribute disproportionately to overall revenue, indicating strategic hubs for Adidas sales.

Regional analysis shows that specific regions dominate revenue and unit sales, highlighting potential for targeted regional campaigns.

Retailer performance varies significantly; top retailers generate most of the revenue and profit, while smaller retailers lag behind.

Sales Channel Insights:

Different sales methods (Online, Outlet, In-Store) show varying performance: In-Store generates the highest revenue, whereas Online drives higher transaction volume.

Analyzing sales by channel helps identify which modes are most profitable versus which generate the most customer engagement.

Overall Strategic Implications:

The data highlights key products, regions, and channels that drive revenue and profit, providing actionable insights for inventory management, marketing strategies, and regional expansion.

Balancing high-margin products with popular high-volume items can optimize overall profitability.

Seasonal patterns suggest opportunities for promotional planning and targeted campaigns in peak months.
